to find columns with missing values or null values



```python
missing_percentages = (df.isna().sum().sort_values(ascending = False)/ len(df))* 100
```

```python
# this will remove all the columns with zero missing values
missing_percentages[missing_percentages!=0]
```

```python
missing_percentages[missing_percentages!=0].plot(kind='barh')
```

Now based on the above graph we can clearly see there are columns with high
missing values so we can avoid using them in our analysis. we can just remove
them using drop() or not use them at all.

### Start Time

```python
df.Start_Time

# we can see the date time is not right format
```

```python
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time[0]
# To check correct datetime format
```

```python
# Checking the accidents by hour of the day
sns.histplot(df.Start_Time.dt.hour, bins = 24, stat='percent')
```

Clearly this graph shows that the accidents happen most from 1 pm to 6 pm
followed by from 6 am to 9 am

6 am to 9 am is generally when office goers travel and from 2 pm to 6 pm is generally the returning time so increase in traffic and accidents

```python
import matplotlib.pyplot as plt

sns.histplot(df.Start_Time.dt.dayofweek, bins = 7, stat='percent')
plt.show()
```

Above graph clearly shows that on weekends the accidents are low as there is drop in number of people going to their offices.


```python
sunday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek==6]
sns.histplot(sunday_start_time.dt.hour, bins = 24)
```

```python
saturday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek==5]
sns.histplot(saturday_start_time.dt.hour, bins=24)
```

On saturday and sunday we can see that the accidents happen on similar timeframe.
Though Saturday has higher number of accidents maybe due to some office goers

The curve of graph shows gaussion curve

```python
Monday_start_time = df.Start_Time[df.Start_Time.dt.dayofweek==0]
sns.histplot(Monday_start_time.dt.hour, bins = 24, legend = True)
```

So cleary when Graphs for Monday and weekends are compared, it shows us that accidents
Here we can see two gaussion curves created due to two peaks

### Months

```python
accidents_by_months = df.Start_Time.dt.month
sns.histplot(accidents_by_months,bins = 12, stat = 'percent')
```

This graph shows accidents by months of all the data i.e. from feb 2016 to dec 2019
also data from jan2016 is missing which will affect the end result.
Also the sources of the data are different which will affect the result.

```python
start_time_2018 = df.Start_Time[df.Start_Time.dt.year == 2018]
sns.histplot(start_time_2018, bins = 12)
```

```python
Severity_level = df.Severity
sns.histplot(Severity_level,bins=6,stat='percent')
```

```python
Severity_level_counts = Severity_level.value_counts()
plt.pie(Severity_level_counts, labels= Severity_level_counts.index,autopct='%.2f')
```
